```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
```

```python
r_consol = 'data/consolidated/'  

# Nombres de los archivos
archivos = [
                'consolidado_alimentos.xlsx',
                'consolidado_combustibles.xlsx',
                'consolidado_educacion.xlsx',
                'consolidado_salud.xlsx',
                'consolidado_vivienda.xlsx'
            ]


# dataframes = {}
# for archivo in archivos:
#     dir = os.path.join(r_consol, archivo)
#     dataframes[archivo] = pd.read_excel(dir)

# # # Mostrar las primeras filas de cada DataFrame
# # for archivo, df in dataframes.items():
# #     print(f"Datos de {archivo}:")
# #     print(df.head())
# #     print("\n")



with pd.ExcelWriter(os.path.join('data/', 'all_data.xlsx')) as writer:
    for archivo in archivos:
        # Obtener el nombre del sector desde el nombre del archivo
        sector = archivo.split('_')[1].replace('.xlsx', '')  # Extrae el sector y elimina la extensi√≥n
        
        # Leer el archivo Excel
        ruta_completa = os.path.join(r_consol, archivo)  # Usa la ruta base r_consol
        df = pd.read_excel(ruta_completa)
        
        # Escribir el DataFrame en una hoja con el nombre del sector
        df.to_excel(writer, sheet_name=sector, index=False)

print("Archivo consolidado creado exitosamente.")

```
