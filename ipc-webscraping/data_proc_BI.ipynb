```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from concurrent.futures import ThreadPoolExecutor
```

```python
r_consol = 'data/consolidated/'  
archivos = [
    'consolidado_alimentos.xlsx',
    'consolidado_combustibles.xlsx',
    'consolidado_educacion.xlsx',
    'consolidado_salud.xlsx',
    'consolidado_vivienda.xlsx',
]

# Funci√≥n para leer y procesar cada archivo
def procesar_archivo(archivo):
    sector = archivo.split('_')[1].replace('.xlsx', '')
    ruta_completa = os.path.join(r_consol, archivo)
    df = pd.read_excel(ruta_completa)
    return (sector, df)

# Usar ThreadPoolExecutor para cargar archivos en paralelo
with ThreadPoolExecutor() as executor:
    resultados = list(executor.map(procesar_archivo, archivos))

# Escribir todos los datos en un solo archivo Excel
with pd.ExcelWriter(os.path.join('data/', 'all_data.xlsx')) as writer:
    for sector, df in resultados:
        df.to_excel(writer, sheet_name=sector, index=False)

print("Archivo consolidado creado exitosamente.")
```
